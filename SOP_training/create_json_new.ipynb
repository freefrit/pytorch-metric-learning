## This code is to create the .json file which content is the dict in shape:

{
'image': filenames,

'label': annotations(num1~12)
}

```python
from google.colab import drive
import os
import json
drive.mount('/content/gdrive')
root = '/content/gdrive/My Drive/Stanford_Online_Products'
```

```python
%cd /content/gdrive/My Drive/Stanford_Online_Products
!ls
```

```python
anno_loc = './annotations/'
if os.path.exists(anno_loc) == False:
  os.mkdir(anno_loc)

train_filenames = []
train_annotations = []

test_filenames = []
test_annotations= []
```

```python
# super_class = [
#       "bicycle", "cabinet", "chair", "coffee_maker", "fan", "kettle", "lamp",
#       "mug", "sofa", "stapler", "table", "toaster"
#   ]

train_path = root + '/Info_Files/Ebay_train.txt'
test_path = root + '/Info_Files/Ebay_test.txt'

print('creat train_img annotations')
f1 = open(train_path, 'r')
f1.readline()
for i in range(0, 59551):
  line = f1.readline()
  s = line.split(' ')
  print(s)
  train_annotations.append(int(s[2]))
  train_filenames.append(s[3].strip('\n'))
f1.close()
print('for training set: ' + str(len(train_filenames)) + ' datas wrote.')

print('creat test_img annotations')
f2 = open(test_path, 'r')
f2.readline()
for i in range(59552, 120054):
  line = f2.readline()
  s = line.split(' ')
  print(s)
  test_annotations.append(int(s[2]))
  test_filenames.append(s[3].strip('\n'))
f2.close()
print('for testing set: ' + str(len(test_filenames)) + ' datas wrote.')  
```

```python
train_filenames
```

```python
train_annot_dict = {
        'image': train_filenames,
        'label': train_annotations
    }
test_annot_dict = {
        'image': test_filenames,
        'label': test_annotations
    }

%cd /content/gdrive/My Drive/Stanford_Online_Products
train_json = json.dumps(train_annot_dict)
f = open('./annotations/OnlineProducts_train_new.json', 'w')
f.write(train_json)
f.close()

test_json = json.dumps(test_annot_dict)
f = open('./annotations/OnlineProducts_test_new.json', 'w')
f.write(test_json)
f.close()

print('annotations have writen to json file')
```
