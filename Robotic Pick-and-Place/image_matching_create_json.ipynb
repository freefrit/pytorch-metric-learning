## This code is to create the .json file which content is the dict in shape:

{
'image': filenames,

'label': annotations(num1~61)
}

```python
from google.colab import drive
import os
import json
drive.mount('/content/gdrive')
root = '/content/gdrive/My Drive/Metric_Learning/image-matching-dataset'
```

```python
%cd /content/gdrive/My Drive/Metric_Learning/image-matching-dataset
!ls
```

```python
anno_loc = './annotations/'
if os.path.exists(anno_loc) == False:
  os.mkdir(anno_loc)

train_filenames = []
train_annotations = []

test_filenames = []
test_annotations= []
```

```python
train_imgs_path = root + '/data/train-imgs.txt'
train_labels_path = root + '/data/train-labels.txt'
test_imgs_path = root + '/data/test-imgs.txt'
test_labels_path = root + '/data/test-labels.txt'

print('creat train_img annotations')
f1 = open(train_imgs_path, 'r')
f2 = open(train_labels_path, 'r')
for i in range(0, 4157):
  s1 = f1.readline()
  print(s1)
  s2 = f2.readline()
  print(s2)
  train_annotations.append(int(s2.strip('\n')))
  train_filenames.append(s1.strip('\n'))
f1.close()
f2.close()
print('for training set: ' + str(len(train_filenames)) + ' datas wrote.')

print('creat test_img annotations')
f1 = open(test_imgs_path, 'r')
f2 = open(test_labels_path, 'r')
for i in range(0, 562):
  s1 = f1.readline()
  print(s1)
  s2 = f2.readline()
  print(s2)
  test_annotations.append(int(s2.strip('\n')))
  test_filenames.append(s1.strip('\n'))
f1.close()
f2.close()
print('for testing set: ' + str(len(test_filenames)) + ' datas wrote.')  
```

```python
train_annot_dict = {
        'image': train_filenames,
        'label': train_annotations
    }
test_annot_dict = {
        'image': test_filenames,
        'label': test_annotations
    }

%cd /content/gdrive/My Drive/Metric_Learning/image-matching-dataset
train_json = json.dumps(train_annot_dict)
f = open('./annotations/image-matching_train_new.json', 'w')
f.write(train_json)
f.close()

test_json = json.dumps(test_annot_dict)
f = open('./annotations/image-matching_test_new.json', 'w')
f.write(test_json)
f.close()

print('annotations have writen to json file')
```
